```python
import os
import requests
from bs4 import BeautifulSoup
import time

# ì €ì¥ í´ë” ìƒì„±
save_path = "pdf_reports"
os.makedirs(save_path, exist_ok=True)

# í¬ë¡¤ë§í•  í˜ì´ì§€ ìˆ˜
total_pages = 281

# User-Agent ì„¤ì • (ë„¤ì´ë²„ ì°¨ë‹¨ ë°©ì§€)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://finance.naver.com/research/debenture_list.naver",
    "Accept-Language": "ko-KR,ko;q=0.9",
    "Connection": "keep-alive",
}

# ìˆ˜ì§‘í•œ ë¦¬í¬íŠ¸ ì •ë³´ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
report_links = []

# 1ï¸âƒ£ ì „ì²´ í˜ì´ì§€ í¬ë¡¤ë§
for page in range(1, total_pages + 1):
    url = f"https://finance.naver.com/research/debenture_list.naver?&page={page}"
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"âŒ í˜ì´ì§€ í¬ë¡¤ë§ ì‹¤íŒ¨: {url}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    
    # í…Œì´ë¸”ì—ì„œ ë¦¬í¬íŠ¸ ëª©ë¡ ì¶”ì¶œ
    rows = soup.select("table.type_1 tr")  # ë„¤ì´ë²„ ë¦¬í¬íŠ¸ ë¦¬ìŠ¤íŠ¸ í…Œì´ë¸” ì„ íƒ
    
    for row in rows[2:]:  # ì²« 2ê°œ í–‰ì€ í—¤ë”, ë¶ˆí•„ìš”í•œ í–‰ì´ë¯€ë¡œ ì œì™¸
        columns = row.find_all("td")
        if len(columns) < 5:  # ë°ì´í„°ê°€ ë¶€ì¡±í•˜ë©´ ê±´ë„ˆëœ€
            continue

        title = columns[0].get_text(strip=True)
        report_page_link = columns[0].find("a")["href"] if columns[0].find("a") else None
        date = columns[3].get_text(strip=True).replace(".", "")  # YYYYMMDD í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        pdf_icon = columns[2].find("a")  # PDF ë‹¤ìš´ë¡œë“œ ë§í¬
        
        if not report_page_link or not pdf_icon:
            continue  # PDF ì—†ëŠ” ë¦¬í¬íŠ¸ ê±´ë„ˆëœ€

        # PDF ë‹¤ìš´ë¡œë“œ URL
        pdf_url = f"https://finance.naver.com{pdf_icon['href']}"
        
        # ë°ì´í„° ì €ì¥
        report_links.append((title, pdf_url, date))

    print(f"âœ… {page}/{total_pages} í˜ì´ì§€ í¬ë¡¤ë§ ì™„ë£Œ")
    time.sleep(1)  # ì„œë²„ ë¶€í•˜ ë°©ì§€

# 2ï¸âƒ£ í¬ë¡¤ë§í•œ ë¦¬í¬íŠ¸ ë§í¬ë¡œ PDF ë‹¤ìš´ë¡œë“œ
for title, pdf_url, date in report_links:
    if len(date) != 8:  # ë‚ ì§œ í˜•ì‹ì´ ì˜¬ë°”ë¥´ì§€ ì•Šë‹¤ë©´ ìŠ¤í‚µ
        print(f"âŒ ì‘ì„±ì¼ ì˜¤ë¥˜ë¡œ íŒŒì¼ ì €ì¥ ê±´ë„ˆëœ€: {title}")
        continue

    response = requests.get(pdf_url, headers=headers)

    if response.status_code == 200:
        pdf_filename = f"{save_path}/{date}_bond.pdf"  # ì‘ì„±ì¼ ê¸°ë°˜ íŒŒì¼ëª… ì„¤ì •

        # PDF ì €ì¥
        with open(pdf_filename, "wb") as f:
            f.write(response.content)

        print(f"âœ… PDF ì €ì¥ ì™„ë£Œ: {pdf_filename}")
    else:
        print(f"âŒ PDF ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨: {pdf_url}")

    time.sleep(2)  # ì„œë²„ ë¶€í•˜ ë°©ì§€

print("ğŸš€ ëª¨ë“  ë¦¬í¬íŠ¸ í¬ë¡¤ë§ ë° ì €ì¥ ì™„ë£Œ!")


```
