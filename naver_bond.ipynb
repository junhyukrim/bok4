```python
import requests
import pdfplumber
import pandas as pd
import io
import time
from bs4 import BeautifulSoup

# 기본 URL
base_url = "https://finance.naver.com/research/debenture_list.naver?page={}"

# 전체 리포트 링크 저장 리스트
report_links = []

# 최대 페이지 설정 (281페이지까지 크롤링)
max_pages = 281

# 페이지 순회하면서 크롤링
for page in range(1, max_pages + 1):
    url = base_url.format(page)
    
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # 채권 리포트 목록 가져오기 (예제: <a href="debenture_read.naver?nid=12345">채권 Daily</a>)
        reports = soup.select("td a[href^='debenture_read.naver']")
        
        for report in reports:
            title = report.text.strip()  # 리포트 제목
            link = "https://finance.naver.com" + report["href"]  # 전체 URL 생성
            
            report_links.append((title, link))
            print(f"📄 제목: {title}")
            print(f"🔗 링크: {link}\n")
    
    else:
        print(f"❌ 페이지 {page} 로딩 실패")

    # 요청 속도 조절 (너무 빠르면 차단될 수 있음)
    time.sleep(1)

print(f"✅ 총 {len(report_links)}개의 리포트 링크 수집 완료!")
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time

# 저장 폴더 생성
save_path = "pdf_reports"
os.makedirs(save_path, exist_ok=True)

# 크롤링할 페이지 수
total_pages = 281

# User-Agent 설정 (네이버 차단 방지)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://finance.naver.com/research/debenture_list.naver",
    "Accept-Language": "ko-KR,ko;q=0.9",
    "Connection": "keep-alive",
}

# 수집한 리포트 정보를 저장할 리스트
report_links = []

# 1️⃣ 전체 페이지 크롤링
for page in range(1, total_pages + 1):
    url = f"https://finance.naver.com/research/debenture_list.naver?&page={page}"
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"❌ 페이지 크롤링 실패: {url}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    
    # 테이블에서 리포트 목록 추출
    rows = soup.select("table.type_1 tr")  # 네이버 리포트 리스트 테이블 선택
    
    for row in rows[2:]:  # 첫 2개 행은 헤더, 불필요한 행이므로 제외
        columns = row.find_all("td")
        if len(columns) < 5:  # 데이터가 부족하면 건너뜀
            continue

        title = columns[0].get_text(strip=True)
        report_page_link = columns[0].find("a")["href"] if columns[0].find("a") else None
        date = columns[3].get_text(strip=True).replace(".", "")  # YYYYMMDD 형식으로 변환
        pdf_icon = columns[2].find("a")  # PDF 다운로드 링크
        
        if not report_page_link or not pdf_icon:
            continue  # PDF 없는 리포트 건너뜀

        # PDF 다운로드 URL
        pdf_url = f"https://finance.naver.com{pdf_icon['href']}"
        
        # 데이터 저장
        report_links.append((title, pdf_url, date))

    print(f"✅ {page}/{total_pages} 페이지 크롤링 완료")
    time.sleep(1)  # 서버 부하 방지

# 2️⃣ 크롤링한 리포트 링크로 PDF 다운로드
for title, pdf_url, date in report_links:
    if len(date) != 8:  # 날짜 형식이 올바르지 않다면 스킵
        print(f"❌ 작성일 오류로 파일 저장 건너뜀: {title}")
        continue

    response = requests.get(pdf_url, headers=headers)

    if response.status_code == 200:
        pdf_filename = f"{save_path}/{date}_bond.pdf"  # 작성일 기반 파일명 설정

        # PDF 저장
        with open(pdf_filename, "wb") as f:
            f.write(response.content)

        print(f"✅ PDF 저장 완료: {pdf_filename}")
    else:
        print(f"❌ PDF 다운로드 실패: {pdf_url}")

    time.sleep(2)  # 서버 부하 방지

print("🚀 모든 리포트 크롤링 및 저장 완료!")


```
