### 오류확인용 + 바로바로 csv 저장

```python
import requests
from bs4 import BeautifulSoup
import time
import random
import csv
find = "기준금리"
# 기존 파일에 이어서 데이터를 추가하고 싶다면 mode='a'
start_date = "2024.01.18"
end_date = "2024.01.22"
# 아.. 12월 31이 지나면 다시 0부터 시작
# 이제야 알았네...ㅠㅠ

page = 1  
articles = []

while True:
    url = f"https://search.naver.com/search.naver?where=news&query={find}&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds={start_date}&de={end_date}&docid=&related=0&mynews=1&office_type=3&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom{start_date.replace('.', '')}to{end_date.replace('.', '')}&is_sug_officeid=0&office_category=3&service_area=0&start={page}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"}
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")
    news_list = soup.select(".news_area")  
    
    
    # 더 이상 기사가 없으면 종료
    if not news_list:
        break

    for news in news_list:
        title_tag = news.select_one(".news_tit")  # 제목 태그
        press_tag = news.select_one(".info.press")  # 신문사 태그
        date_tag = news.select_one("span.info")  # 날짜 태그 
        
        if title_tag and press_tag and date_tag:
            title = title_tag.text.strip()  # 기사 제목
            link = title_tag["href"]  # 기사 링크
            news_agency = press_tag.text.strip()  # 신문사 이름
            article_date = date_tag.text.strip()  # 날짜
            
            articles.append((title, news_agency, article_date, link))
        with open('test.csv', mode='a',newline='',encoding ='utf-8') as file:
            writer = csv.writer(file)
            writer.writerows(articles)
    print(f"{page}페이지크롤링, 날짜: {article_date}")
    # 돌려보니 기사 개수 보단 날짜를 출력해서 남은기간 알수있게 하는게 더좋아보임
    
    # 다음 페이지로 이동
    page += 10  # 10개씩 넘어감 
    sleep_time = random.uniform(1,2)
    time.sleep(sleep_time)
for idx,(title, news_agency, article_date, link) in enumerate(articles):
    print(f"'{article_date}','{news_agency}','{title}', '{link}'")
with open('test.csv', mode='a',newline='',encoding ='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(articles)
```

```python
import random
import requests
from bs4 import BeautifulSoup
import time
import csv

# 여러 User-Agent를 리스트로 저장
user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Edge/91.0.864.59 Safari/537.36"
    ]

find = "기준금리"
start_date = "2024.04.10"
end_date = "2024.12.31"

page = 1  
articles = []
# 이미 크롤링한 기사들의 링크를 추적하기 위한 set
crawled_links = set()
while True:
    # 랜덤으로 User-Agent 선택
    headers = {"User-Agent": random.choice(user_agents)}

    url = f"https://search.naver.com/search.naver?where=news&query={find}&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds={start_date}&de={end_date}&docid=&related=0&mynews=1&office_type=3&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom{start_date.replace('.', '')}to{end_date.replace('.', '')}&is_sug_officeid=0&office_category=3&service_area=0&start={page}"

    res = requests.get(url, headers=headers)
    
    if res.status_code != 200:  # 응답 코드가 200이 아니면 차단된 것으로 간주
        print(f"응답 코드 오류: {res.status_code}. 크롤링 중지.")
        break
    
    soup = BeautifulSoup(res.text, "html.parser")
    news_list = soup.select(".news_area")

    if not news_list:
        print("뉴스 목록이 비어있습니다. 차단되었을 가능성이 있습니다.")
        break

    for news in news_list:
        title_tag = news.select_one(".news_tit")
        press_tag = news.select_one(".info.press")
        date_tag = news.select_one("span.info")
        
        if title_tag and press_tag and date_tag:
            title = title_tag.text.strip()
            link = title_tag["href"]
            news_agency = press_tag.text.strip()
            article_date = date_tag.text.strip()
            if link in crawled_links:
                print(f"중복된 기사 발견, 무시합니다: {title}")
                continue

            # 기사 데이터를 리스트에 추가하고 바로 CSV 파일에 저장
            with open('article_data1.csv', mode='a', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow((title, news_agency, article_date, link))

    print(f"{page}페이지 크롤링 완료, 진행상황 : {article_date},반복확인{title}")
    page += 10
    sleep_time = random.uniform(1, 5)
    time.sleep(sleep_time)
```
