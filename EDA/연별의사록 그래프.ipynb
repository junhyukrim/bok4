```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# CSV 파일 읽기 (파일 경로는 실제 경로로 변경하세요)
df = pd.read_csv('연별의사록.csv')
# 2025는 1개 여서 뺐습니다. 


# seaborn 스타일 적용
sns.set(style="whitegrid")
year = df['year']
# 그래프 그리기
plt.figure(figsize=(8, 6))  # 그래프 크기 설정
plt.bar(df['year'], df['total'], color='#B0E0E6')#'skyblue','red' 이런거넣으면 됨

# 그래프 제목과 레이블 설정
plt.title('Total count of minutes', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.xticks(year, rotation=45)
plt.ylabel('Total', fontsize=12)

# X축 눈금 설정 
plt.xticks(df['year'])

# 그래프 표시
plt.grid(True)
plt.show()
```
