```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일을 읽어옵니다.
df = pd.read_csv('연_언론사별 기사갯수.csv')  # 실제 파일 경로를 입력하세요.

# Year을 X축, 각 언론사 별 데이터를 Y축으로 설정
years = df['Year']
financial = df['Financial']
money_today = df['MoneyToday']
edaily = df['E-Daily']
asia_economy = df['AsiaEconomy']
total = df['Total']

# seaborn 스타일 적용
sns.set(style="whitegrid")



# 막대그래프 그리기
plt.figure(figsize=(10, 6))

plt.bar(years, financial, label='Financial', color='#B0E0E6')  # 파스텔 파란색
plt.bar(years, money_today, bottom=financial, label='Money Today', color='#98FB98')  # 파스텔 초록색
plt.bar(years, edaily, bottom=financial + money_today, label='E-Daily', color='#FFB6C1')  # 파스텔 빨간색
plt.bar(years, asia_economy, bottom=financial + money_today + edaily, label='Asia Economy', color='#FFFACD')  # 파스텔 노란색



# 그래프 제목과 레이블 추가
plt.xlabel('Year')
plt.ylabel('Total')
plt.title('Number of News Articles by Year')
plt.xticks(years, rotation=45)
plt.legend()

# 그래프 PNG로 저장
plt.tight_layout()
plt.savefig('stacked_news_coverage_graph.png', format='png')  # 원하는 파일 이름과 형식으로 저장

# 그래프 표시
plt.show()
```
